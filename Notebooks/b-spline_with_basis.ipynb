```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as si
```

```python
points = [[0.000, 0.5],
          [0.125, 0.8],
          [0.375, -0.5],
          [0.625, 0.0],
          [0.875, 1.0],
          [1.000, 0.5]];
points = np.array(points)
x = points[:,0]
y = points[:,1]

degree = 3
```

```python
t = range(len(points))
ipl_t = np.linspace(0.0, len(points) - 1, 50)

x_tup = si.splrep(t, x, k=degree)
x_list = list(x_tup)
y_tup = si.splrep(t, y, k=degree)
y_list = list(y_tup)

xl = x.tolist()
x_list[1] = xl

yl = y.tolist()
y_list[1] = yl

x_i = si.splev(ipl_t, x_list)
y_i = si.splev(ipl_t, y_list)
```

```python
fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(211)
plt.xlim([min(x) - 0.05, max(x) + 0.05])
plt.ylim([min(y) - 0.05, max(y) + 0.05])

plt.plot(x, y, 'r--', lw=1)
plt.scatter(x, y, c='g')
plt.plot(x_i, y_i, 'b', lw=2)

ax = fig.add_subplot(212)
for i in range(len(points)):
    vec = np.zeros(len(points)+degree+1)
    vec[i] = 1.0
    x_list[1] = vec.tolist()
    x_i = si.splev(ipl_t, x_list)
    plt.plot(ipl_t, x_i)
plt.xlim([0.0, max(t)])
plt.show()
```
